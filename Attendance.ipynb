```python
import csv

CSV_FILE = "attendance.csv"

def load_csv(filename):
    with open(filename, newline="") as f:
        return list(csv.reader(f))

def save_csv(filename, rows):
    with open(filename, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerows(rows)

def update_attendance():
    rows = load_csv(CSV_FILE)

    header = rows[0]
    people = rows[1:]

    # Add new column label: next training number
    new_training = f"session {len(header)}"
    header.append(new_training)

    print(f"Recording attendance for {new_training}...\n")

    for row in people:
        name = row[0]
        answer = input(f"Was {name} present? (y/n): ").strip().lower()

        if answer == "y":
            row.append("1")
        else:
            row.append("0")

    save_csv(CSV_FILE, [header] + people)
    print("\nAttendance updated. CSV saved.")

if __name__ == "__main__":
    update_attendance()
```

```python
attendance_list = {}

with open("attendance.csv", newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        name = row["name"]
        attendance_list[name] = [bool(int(v)) for k, v in row.items() if k != "name"]

total_sessions = len(next(iter(attendance_list.values())))
```

```python
def presence_checker(attendance_pp : dict[str, list[bool]]) -> dict[str, int]:
    total_attendance_pp = {}

    for name, attendances in attendance_pp.items():
        total_attendance_pp[name] = sum(attendances)

    return total_attendance_pp
```

```python
what_to_check = input("For who would you like to check the total attendance? 1 name, or 'all' for everyone").lower().strip()

if what_to_check == "all":
    sorting = input("Would you like to print the list in order from high to low? y/n")
    print(f"In total there have been {total_sessions} sessions")

    attendance_dict = presence_checker(attendance_list)
    if sorting == "y":
        nr = 1

        for name, attendance in sorted(presence_checker(attendance_list).items(), key=lambda x: x[1], reverse=True):
            print(f"{nr}: {name.capitalize()} \t has been present {attendance} times \t That is {attendance / total_sessions * 100:.1f}% ")
            nr += 1
    else:
        for naam, aanwezigheid in presence_checker(attendance_list).items():
            print(f"{naam.capitalize()} \t is {aanwezigheid} keer aanwezig geweest \t Dat is {aanwezigheid / total_sessions * 100:.1f}% ")


elif what_to_check in presence_checker(attendance_list):
    print(f"{what_to_check.capitalize()} has been present {presence_checker(attendance_list)[what_to_check]} times, that is {presence_checker(attendance_list)[what_to_check] / total_sessions * 100:.1f}%")

else:
    print(f"{what_to_check.capitalize()} has not been found in the attendance list.")
```
